# Matrix Dimensions and Entries

## Matrix Dimensions
A matrix is often described by the number of rows & columns that it has.  For instance, a 3 x 4 matrix is a matrix with 3 rows and 4 columns.  In the description "3 x 4", the number of rows always comes first, and the numbers of columns always comes second, so remember:

"rows x columns"

A matrix can be as small as 1 x 1, with one row and one column, in which case it looks like this:

$[a]$

Or it can have infinitely many rows and/or columns.  It can have the same number of rows and columns, more rows than columns, or more columns than rows. 

A $2\times2$ matrix: 
$
  \left[\begin{array}{rr}
  a & b \\
  c & d \\
  \end{array}\right]
$

A $2\times3$ matrix: 
$
  \left[\begin{array}{rrr}
  a & b & c \\
  d & e & f \\
  \end{array}\right]
$

A $3\times2$ matrix: 
$
  \left[\begin{array}{rr}
  a & b \\
  c & d \\
  e & f \\
  \end{array}\right]
$

Matrices are just a way of arranging data.

Example

Give the dimensions of each matrix.

$A=\left[\begin{array}{rrr}
  7 & 3 & 4 \\
  1 & 6 & 1 \\
  2 & 2 & 3 \\
  \end{array}\right]$

$B=\left[\begin{array}{rrrr}
  1 & 9 & 0 & 0 & 2 \\
  \end{array}\right]$

$C=\left[\begin{array}{r}
  0 \\
  0 \\
  1 \\
  \end{array}\right]$

Matrix A is a $3\times3$ matrix.

Matrix B is a $1\times5$ matrix.

Matrix C is a $3\times1$ matrix.

## Matrix Entries

We use the name of the matrix & the row & column where the entry is.

Example

  k_{1,1} & k_{1,2} & k_{1,3} \\
